Here is a Python script that performs a wavelet transformation using the `PyWavelets` library. This example demonstrates how to perform a Discrete Wavelet Transform (DWT) on a signal and then reconstruct the signal from the wavelet coefficients.

First, make sure you have the PyWavelets library installed. You can install it using pip:
```
pip install PyWavelets
```
Here is a Python script for wavelet transformation:




````python
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Generate a sample signal: a combination of two sinusoids
t = np.linspace(0, 1, 400)
signal = np.sin(2 * np.pi * 7 * t) + np.sin(2 * np.pi * 13 * t)

# Perform Discrete Wavelet Transform (DWT)
wavelet = 'db1'  # Daubechies 1 wavelet
coeffs = pywt.wavedec(signal, wavelet)

# Plot the original signal
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, signal, label='Original Signal')
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

# Plot the wavelet coefficients
for i, coeff in enumerate(coeffs):
    print(i)
    print(coeff)
    plt.subplot(3, 1,  1+i)
    plt.plot(coeff, label=f'Detail Coefficients Level {i}')
    plt.title(f'Detail Coefficients Level {i}')
    plt.xlabel('Samples')
    plt.ylabel('Amplitude')
    plt.legend()

plt.tight_layout()
plt.show()

# Signal reconstruction
reconstructed_signal = pywt.waverec(coeffs, wavelet)

# Plot the reconstructed signal
plt.figure(figsize=(12, 4))
plt.plot(t, reconstructed_signal, label='Reconstructed Signal')
plt.title('Reconstructed Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
````
